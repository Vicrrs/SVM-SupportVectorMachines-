# Aplicando Label Encoder

Iremos aplicar o Label Encoder porque o SVM irá se basear em valores numéricos para definir o melhor plano, então ele só sabe trabalhar com variáveis numéricas. Iremos transformar a coluna Purpose em numérica para continuar.

```python
dados.dtypes
```

```python
dados.dtypes.reset_index()
```

```python
colunas = dados.dtypes.reset_index()
```

```python
colunas.columns
```

```python
colunas[colunas[0] == 'object']['index']
```

```python
categ_col = colunas[colunas[0] == 'object']['index'].to_list()
```

```python
categ_col
```

```python
# Fazendo o Label Encoder
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
```

```python
#criando  um coluna purpose encoded
for i in categ_col:
  le.fit(dados[i])
  dados[str(i) + '_encoded'] = le.transform(dados[i])
```

```python
dados
```

```python
# Eliminando a coluna que nao preciso mais (purpose)
dados = dados.drop('purpose', axis=1)
```

```python
dados.head()
```

# Implementando o modelo

Variável que queremos prever é a Not Fully Paid (ou seja, se a pessoa não pagou totalmente o impréstimo)

## Separando a base em X (variáveis de entrada) e Y (variável que queremos prever)

```python
#separando o dataframe em x -> features/variaveis de entrada, y -> variaveis de resposta/target
x = dados.drop('not.fully.paid', axis=1)
y = dados['not.fully.paid']
```

## Quebrando o dataset em treino e teste - Usando 70% para treino e 30% para teste

```python
from sklearn.model_selection import train_test_split

x_train, x_test , y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=42)
```

## Carregando o Modelo

```python
from sklearn.svm import SVC

# Selecionando qual modelo eu quero treinar, manteremos o kernel linear
# mas importante pesquisar sobre outros tipos de kernel para o SVM (sigmoid, radial  e polynomial)

svm = SVC(kernel='linear')
svm.fit(x_train, y_train)

# Fazendo a predição

y_pred = svm.predict(x_test)
```

## Avaliando resultado

```python
from sklearn.metrics import accuracy_score

print(accuracy_score(y_test, y_pred))
```

## Conclusão

Dataset de emprestismos, vamos prever a probabilidade de uma pessoal pagar ou nao um emprestimo. Conseguimos um resultado de 82%, para a possibilidade de ela pagar o empréstimo.
